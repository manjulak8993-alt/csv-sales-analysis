```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("sales_data.csv")  
print("First 5 rows:\n", df.head())
print("\nData Info:\n")
print(df.info())
print("\nSummary Statistics:\n", df.describe())
sales_by_product = df.groupby("Product")["Sales"].sum()
sales_by_region = df.groupby("Region")["Sales"].sum()
print("\nTotal Sales by Product:\n", sales_by_product)
print("\nTotal Sales by Region:\n", sales_by_region)
sales_by_product.plot(kind="bar", title="Total Sales by Product", figsize=(6,4))
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
sales_by_region.plot(kind="bar", title="Total Sales by Region", figsize=(6,4))
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```
